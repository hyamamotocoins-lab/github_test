# 00 — Paperspace project setup

このノートブックを同じプロジェクト内の全段階に先立って一度実行します。最初のコードセルは、唯一不足している想定の `pytest` だけを導入します。PyTorch/CUDA や NumPy は既存の Paperspace 環境を変更しません。

```python
# 必ず最初に pytest を用意する。
import importlib.util
import subprocess
import sys

if importlib.util.find_spec('pytest') is None:
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'pytest>=8'])
import pytest
print('pytest:', pytest.__version__)
```

```python
from pathlib import Path
import os
import sys

BUNDLE_NAME = 'validated_4d_su2_rg_codex_bundle'
explicit = os.environ.get('VALIDATED_RG_PROJECT_ROOT')
candidates = ([Path(explicit)] if explicit else []) + [
    Path.cwd(), Path.cwd().parent, Path.cwd() / BUNDLE_NAME,
    Path('/notebooks') / BUNDLE_NAME, Path('/storage') / BUNDLE_NAME,
]
PROJECT_ROOT = next((p.expanduser().resolve() for p in candidates
                     if (p.expanduser() / 'validated_4d_su2_rg_codex_design.md').is_file()), None)
if PROJECT_ROOT is None:
    raise RuntimeError('プロジェクト全体を配置したディレクトリを検出できません。')
if sys.version_info < (3, 11):
    raise RuntimeError(f'Python 3.11+ が必要です: {sys.version.split()[0]}')
PERSIST_ROOT = Path(os.environ.get('VALIDATED_RG_PERSIST_ROOT', '/storage/validated_4d_su2_rg')).expanduser().resolve()
os.environ['VALIDATED_RG_PROJECT_ROOT'] = str(PROJECT_ROOT)
os.environ['VALIDATED_RG_PERSIST_ROOT'] = str(PERSIST_ROOT)
os.environ['VALIDATED_RG_PERSIST_ACK'] = 'I_CONFIRM_THIS_PATH_IS_PERSISTENT'
for path in (PERSIST_ROOT, PROJECT_ROOT / 'src', PROJECT_ROOT / 'tests'):
    path.mkdir(parents=True, exist_ok=True)
print('PROJECT_ROOT =', PROJECT_ROOT)
print('PERSIST_ROOT =', PERSIST_ROOT)
print('次のノートブックも同じ project と /storage を使用してください。')
```

```python
import platform
import numpy as np

diagnostics = {
    'python': sys.version.split()[0],
    'platform': platform.platform(),
    'pytest': pytest.__version__,
    'numpy': np.__version__,
}
try:
    import torch
    diagnostics.update({
        'torch': torch.__version__,
        'cuda_available': torch.cuda.is_available(),
        'cuda_version': torch.version.cuda,
        'gpu': torch.cuda.get_device_name(0) if torch.cuda.is_available() else None,
    })
except ImportError:
    diagnostics['torch'] = None
diagnostics
```
